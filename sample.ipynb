```python
import pandas as pd
```

```python
df=pd.read_csv("D:\project 2 predictive\Reviews.csv")
```

```python
print(df.head())
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# Convert rating (Score) → sentiment label
def convert_label(rating):
    if rating <= 2:
        return 0   # Negative
    elif rating == 3:
        return 1   # Neutral
    else:
        return 2   # Positive

df["label"] = df["Score"].apply(convert_label)
```

```python
# Keep only required columns
df = df[["Text", "label"]]

# Rename Text → text (for consistency)
df = df.rename(columns={"Text": "text"})

# Remove missing values
df = df.dropna()
```

```python
df = df.sample(5000, random_state=42)
```

```python

from sklearn.model_selection import train_test_split

# Split data
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Extract text and labels
train_texts = train_df["text"].tolist()
test_texts = test_df["text"].tolist()

train_labels = train_df["label"].tolist()
test_labels = test_df["label"].tolist()
```

```python
print("Train size:", len(train_texts))
print("Test size:", len(test_texts))
```

```python
from collections import Counter

print("Train labels:", Counter(train_labels))
print("Test labels:", Counter(test_labels))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
vectorizer = TfidfVectorizer(max_features=5000)

X_train = vectorizer.fit_transform(train_texts)
X_test = vectorizer.transform(test_texts)
```

```python
svm_model = SVC(kernel='linear')

svm_model.fit(X_train, train_labels)
```

```python
svm_preds = svm_model.predict(X_test)
```

```python

svm_acc = accuracy_score(test_labels, svm_preds)

print("SVM Accuracy:", svm_acc)
```

```python

print("\nClassification Report:\n")
print(classification_report(test_labels, svm_preds))
```
